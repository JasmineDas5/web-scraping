```python
from bs4 import BeautifulSoup
```

```python
df=pd.DataFrame(columns=['Book_name','author_name','Rating','Book_type','Book_price','Book_link'])
def get_data(pageno):
  url='https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageno)+'_books?ie=UTF8&pg='+str(pageno)  #where there is pg=1 there u write page no in str not 1
  response=requests.get(url)
  soup=BeautifulSoup(response.content,'html.parser')
  for d in soup.find_all('div',attrs={"class":"zg-grid-general-faceout"}):          # the whole div ,where it starts
    book_name=d.find('div',attrs={"class":"_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"})       # the div where book is present
    author_name=d.find('div',attrs={"class":"a-row a-size-small"})
    rating=d.find('span',attrs={"class":"a-icon-alt"})
    book_type=d.find('span',attrs={"class":"a-size-small a-color-secondary a-text-normal"})
    price=d.find('span',attrs={"class":"a-size-base a-color-price"})
    book_link=d.find('a',attrs={"class":"a-link-normal"})
    # print(book_name.text)
    # print(author_name.text)
    # print(rating.text)
    # print(book_type.text)
    # print(price.text)
    # print("https:www.amazon.in/"+book_link.get('href'))     # as anchor dont have text u have to get it
    all=[]
    if book_name is not None:
      all.append(book_name.text)
    else:
      all.append(" ")
    if author_name is not None:
      all.append(author_name.text)
    else:
      all.append(" ")
    if rating is not None:
      all.append(rating.text)
    else:
      all.append(" ")
    if book_type is not None:
      all.append(book_type.text)
    else:
      all.append(" ")
    if price is not None:
      all.append(price.text)
    else:
      all.append(" ")
    if book_link is not None:
      all.append('https:/www.amazon.in'+book_link.get('href'))
    else:
      all.append(" ")
    df.loc[len(df)]=all
  return df

```
